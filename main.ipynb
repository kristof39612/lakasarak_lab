--> Van olyan hely, ahol a ```postcode``` üres, de a ```city``` kitöltött
- Kipótolhatjuk ```city``` alapján pl Budapest IV. > 1041 | itt az utolsó szám az nem lesz valós, hiszen az változhat, de a ```postcode``` cserébe nem üres
- Kipótolhatjuk ```city``` alapján, de úgy, hogy az 1AAB ```postcode``` számsorozatból az utolsó számjegy jelezheti, hogy itt nem tudjuk a pontos adatot, de a kerületet igen.

--> Elsőnek nézzük meg, hogy a ```postcode``` az helyes-e ott, ahol tudjuk a kerületet.

````python
citydf = df[df['city'].notnull()]
citydf = citydf[citydf['postcode'].notnull()]
citydf.sort_values(by='city', ascending=True, inplace=True)
citydf

def is_postcode_correct(city, postcode):
    match = re.match(r'Budapest (\w+)\.', city)
    if match:
        roman_numeral = match.group(1)
        decimal_number = str(roman.fromRoman(roman_numeral))
        if len(decimal_number) == 1:
            decimal_number = '0' + decimal_number
        return str(postcode).startswith('1' + decimal_number)
    return False

citydf['postcode_correct'] = citydf.apply(lambda row: is_postcode_correct(row['city'], row['postcode']), axis=1)
citydf
````

````python
citydf['postcode_correct'].value_counts()
````

--> **211 ```postcode``` nem jó**

````python
incorrect_postcodes = citydf[citydf['postcode_correct'] == False]
incorrect_postcodes
````

--> #TODO ezekkel valamit kezdeni, **mivel kevés érték ezért droppolni is lehetne őket**

--> TFH, hogy a ```postcode``` jó, írjuk vissza a fő dataframe-be a megfelelő kerületet

````python
def get_district_from_postcode(postcode):
    district_number = int(str(postcode)[1:3])
    return roman.toRoman(district_number)

def update_city(row):
    if not row['postcode_correct']:
        district_roman = get_district_from_postcode(row['postcode'])
        return f"Budapest {district_roman}."
    return row['city']

citydf['city'] = citydf.apply(update_city, axis=1)
citydf
````

````python
citydf['postcode_correct'] = citydf.apply(lambda row: is_postcode_correct(row['city'], row['postcode']), axis=1)
citydf
````

````python
citydf['city'].unique()
````

````python
citydf.drop(columns=['postcode_correct'],inplace=True)
````

````python
unknown_postcodes = df[df['postcode'].isna() & df['city'].notna()]
````

````python
unknown_postcodes
````

````python
def fill_postcode_from_city(row):
    match = re.match(r'Budapest (\w+)\.', row['city'])
    if match:
        roman_numeral = match.group(1)
        decimal_number = str(roman.fromRoman(roman_numeral))
        if len(decimal_number) == 1:
            decimal_number = '0' + decimal_number
        return float(f"1{decimal_number}1")
    return row['postcode']

unknown_postcodes['postcode'] = unknown_postcodes.apply(fill_postcode_from_city, axis=1)
unknown_postcodes
````
