--> Van olyan hely, ahol a ```postcode``` üres, de a ```city``` kitöltött
- Kipótolhatjuk ```city``` alapján pl Budapest IV. > 1041 | itt az utolsó szám az nem lesz valós, hiszen az változhat, de a ```postcode``` cserébe nem üres
- Kipótolhatjuk ```city``` alapján, de úgy, hogy az 1AAB ```postcode``` számsorozatból az utolsó számjegy jelezheti, hogy itt nem tudjuk a pontos adatot, de a kerületet igen.

--> Elsőnek nézzük meg, hogy a ```postcode``` az helyes-e ott, ahol tudjuk a kerületet.

````python
citydf = df[df['city'].notnull()]
citydf = citydf[citydf['postcode'].notnull()]
citydf.sort_values(by='city', ascending=True, inplace=True)
citydf

def is_postcode_correct(city, postcode):
    match = re.match(r'Budapest (\w+)\.', city)
    if match:
        roman_numeral = match.group(1)
        decimal_number = str(roman.fromRoman(roman_numeral))
        if len(decimal_number) == 1:
            decimal_number = '0' + decimal_number
        return str(postcode).startswith('1' + decimal_number)
    return False

citydf['postcode_correct'] = citydf.apply(lambda row: is_postcode_correct(row['city'], row['postcode']), axis=1)
citydf
````

````python
citydf['postcode_correct'].value_counts()
````

--> **211 ```postcode``` nem jó**

````python
citydf = citydf[citydf['postcode_correct'] == False]
citydf
````
